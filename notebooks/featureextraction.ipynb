```python
import re
import emoji

r = re.compile('|'.join(list(re.escape(p) for p in emoji.UNICODE_EMOJI.keys())))

def find_emojis(a_list):

    aux = [' '.join(r.findall(s)) for s in a_list]
    aux = set(list(filter(None, aux)))
    return(aux)

def clean_text(text):
    '''
    очистка текста
    
    на выходе очищеный текст
    
    '''
   
    if not isinstance(text, str):
        text = str(text)
    
    text = text.lower()
    text = text.strip('\n').strip('\r').strip('\t')

    text = re.sub("-\s\r\n\|-\s\r\n|\r\n", '', str(text))

    text = re.sub("[0-9]|[-—,:;_%©«»*@#№$^•·&()]|[+=]|[[]|[]]|[/]|", '', text)
    text = re.sub(r"\r\n\t|\n|\\s|\r\t|\\n", ' ', text)
    text = re.sub(r'[\xad]|[\s+]', ' ', text.strip())
    
    emjs = find_emojis(text)
    for emj in emjs:
        text = text.replace(emj, ' ' + emj + ' ')

    return text

clean_text(text)
```

```python
cache = {}

def lemmatization(text, delimiters):
    '''
    лемматизация
        [0] если зашел тип не `str` делаем его `str`
        [1] токенизация предложения через razdel
        [2] проверка есть ли в начале слова '-'
        [3] проверка токена с одного символа
        [4] проверка есть ли данное слово в кэше
        [5] лемматизация слова
        [6] проверка на стоп-слова

    на выходе лист отлемматизированых токенов
    '''

    # [0]
    if not isinstance(text, str):
        text = str(text)
    
    # [1]
    tokens = list(tokenizer(text))
    words = [_.text for _ in tokens]

    words_lem = []
    for i, w in enumerate(words):
        
        if w[0] == '-' or w[0] == '–': # [2]
            w = w[1:]
        else: # [3]
            if w in cache: # [4]
                words_lem.append(cache[w])
            else: # [5]
                if w[-1] in delimiters:
                    w = w.strip(delimiters)
                    temp_cach = cache[w] = morph.parse(w)[0].normal_form+words[i][-1]
                    words_lem.append(temp_cach)
                else:
                    temp_cach = cache[w] = morph.parse(w)[0].normal_form
                    words_lem.append(temp_cach)
    
#     words_lem_without_stopwords=[i for i in words_lem if not i in stopword_ru] # [6]
    
    return [i for i in words_lem if i and i!=' '] #words_lem_without_stopwords
```

```python
%%time
# features
nodes = []
edges = []
ATD = [] # average total degree
LCC = [] # largest connected component
LSC = [] #largest strongly connected component
L1 = [] # number of loops with one
L2 = [] # two 
L3 = [] # or three nodes
Density = []
Diameter = []
ASP = [] # average shortest path


num = 0
for i, lemmas in zip(data.index[12:],data.lemmas.astype(str).tolist()[12:]):
# for i, lemmas in enumerate(data.lemmas.astype(str).tolist()):
    G = cooccurrence_net(lemmas.split(), ['.','!','?'], d="directed",
                     w="unweighted", window=1, lower="No")
    print('matrix done')
    
    nodes.append(G.number_of_nodes())
    data.set_value(i, "Nodes", nodes[-1])
    print('nodes done')
    edges.append(G.number_of_edges())
    data.set_value(i, "Edges", edges[-1])
    print('edges done')
    ATD.append(np.mean([i[-1] for i in G.in_degree()]))
    data.set_value(i, "ATD", ATD[-1])
    print('atd')
    G_ = G.to_undirected()
    mG = max(nx.connected_component_subgraphs(G_), key=len)
    LCC.append(len(mG))
    data.set_value(i, "LCC", LCC[-1])
    print('lcc')
    sG = max(nx.strongly_connected_components(G), key=len)
    LSC.append(len(sG))
    data.set_value(i, "LSC", LSC[-1])
    print('lsc')
    sc = nx.simple_cycles(G)
    print('simple cycles')
    lens = [len(e) for e in sc]
    L1.append(len([e for e in lens if e==1]))
    data.set_value(i, "L1", L1[-1])
    print('l1')
    
    L2.append(len([e for e in lens if e==2]))
    data.set_value(i, "L2", L2[-1])
    L3.append(len([e for e in lens if e==3]))
    data.set_value(i, "L3", L3[-1])
    Density.append(nx.density(G))
    data.set_value(i, "Density", Density[-1])
    print('density')
    Diameter.append(nx.diameter(mG))
    data.set_value(i, "Diameter", Diameter[-1])
    print('diameter')
    ASP.append(np.mean([nx.average_shortest_path_length(g) for g in nx.connected_component_subgraphs(G_)]))
    data.set_value(i, "ASP", ASP[-1])
    print('asp')
    data.to_csv('PsyHack_RUDN_test.csv', sep='\t', index=False)
#     print('saved')
    num+=1
    print(num)
```

```python
colors = cl.scales['12']['qual']['Paired']

def scatter_nodes(pos, degrees, nodeID, part=None, labels=None, color='#a3a3c2', size=8, opacity=1):
    # pos is the dict of node positions
    # labels is a list  of labels of len(pos), to be displayed when hovering the mouse over the nodes
    # color is the color for nodes. When it is set as None the Plotly default color is used
    # size is the size of the dots representing the nodes
    # opacity is a value between [0,1] defining the node color opacity

    trace = Scatter(x=[], 
                    y=[],  
                    mode='markers', 
                    marker=Marker(
        showscale=False,
        colorscale='Greens',
        reversescale=True,
        color=[], 
        size=[],
    line=dict(width=0)))
    for nd in nodeID:
        trace['x'].append(pos[nd][0])
        trace['y'].append(pos[nd][1])
        trace['marker']['size'].append(degrees[nd])#1000*(degrees[nd]/sum(list(degrees.values())))) #degrees[nd])
        if part:
            col = colors[part[nd] % len(colors)]
            trace['marker']['color'].append(col)
        
    attrib=dict(name='', text=labels , hoverinfo='text', opacity=opacity) # a dict of Plotly node attributes
    trace=dict(trace, **attrib)# concatenate the dict trace and attrib
    if color:
        trace['marker']['color']=color
#     trace['marker']['size']=size

    return trace  

def scatter_edges(G, pos, line_color='#a3a3c2', line_width=1, opacity=.2):
    trace = Scatter(x=[], 
                    y=[], 
                    mode='lines'
                   )
    for edge in G.edges():
        trace['x'] += [pos[edge[0]][0],pos[edge[1]][0], None]
        trace['y'] += [pos[edge[0]][1],pos[edge[1]][1], None]  
        trace['hoverinfo']='none'
        trace['line']['width']=line_width
        if line_color is not None: # when it is None a default Plotly color is used
            trace['line']['color']=line_color
    return trace       
```

```python
def draw(title, trace1, trace2):
    width=600
    height=600
    axis=dict(showline=False, # hide axis line, grid, ticklabels and  title
              zeroline=False,
              showgrid=False,
              showticklabels=False,
              title='' 
              )
    layout=Layout(title= '',

        font= Font(),
        showlegend=False,
        autosize=False,
        width=width,
        height=height,
        xaxis=dict(
            title=title,
            titlefont=dict(
            size=16,
            color='#a3a3c2'),
            showgrid=False,

            showline=False,
            showticklabels=False,
            zeroline=False
        ),
        yaxis=YAxis(axis),
        margin=Margin(
            l=40,
            r=40,
            b=85,
            t=100,
            pad=0,

        ),
        hovermode='closest',
        paper_bgcolor='rgba(0,0,0,1)',
        plot_bgcolor='rgba(0,0,0,1)'
        )


    data=Data([trace1, trace2])

    fig = Figure(data=data, layout=layout)
    py.iplot(fig, image='png')
```
